```python
from src.neural_decoder.rnn_trainer import trainModel

for bs in [32, 64]:
    for lr in [0.02, 0.03]:
        for un in [512, 1024]:
            for nl in [3, 5]:
                modelName = 'speechLSTMw/oLayerNorm' + '_batchSize' + str(bs) + '_lr' + str(lr) + '_nUnits' + str(un) + '_nLayers' + str(nl)
                print(modelName)

                args = {}
                args["model_type"] = "LSTM"
                args['outputDir'] = '/content/drive/MyDrive/Colab Notebooks/ECSE 552/Project/speech_logs/' + modelName
                args['datasetPath'] = '/content/drive/MyDrive/Colab Notebooks/ECSE 552/Project/ptDecoder_ctc'
                args['seqLen'] = 150
                args['maxTimeSeriesLen'] = 1200
                args['batchSize'] = bs
                args['lrStart'] = lr
                args['lrEnd'] = lr
                args['nUnits'] = un #1024
                args['nBatch'] = 10000
                args['nLayers'] = nl #5
                args['seed'] = 0
                args['nClasses'] = 40
                args['nInputFeatures'] = 256
                args['dropout'] = 0.4
                args['whiteNoiseSD'] = 0.8
                args['constantOffsetSD'] = 0.2
                args['gaussianSmoothWidth'] = 2.0
                args['strideLen'] = 4
                args['kernelLen'] = 32
                args['bidirectional'] = True
                args['l2_decay'] = 1e-5

                args['earlyStop_patience'] = 10
                args['earlyStop_minChange'] = 0.005
                args['nEpochs'] = 100
                args['gradClipVal'] = 10

                run = wandb.init(
                    project="ECSE552-Project-LSTM",
                    name=modelName,
                    config=args,
                )

                trainModel(args)
```
